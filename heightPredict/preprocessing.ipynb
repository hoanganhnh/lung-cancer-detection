```python
# Normalize các chỉ số như Father, Mother
# min với Min là 0 ( = 0cm), Max là 100 ( = 200cm) trong khoảng 0 - 1
# kết quả lưu ở file 1. normalized
# chuyển sang dạng Binary các giá trị Kids và Gender
# kết quả lưu ở file 2. toBinary
```

```python
df = pd.read_csv("3. normalized.csv")
```

```python
def normalize(values):
    min_value = min(values)
    max_value = max(values)

    normalized_values = []
    for value in values:
        normalized_value = (value - min_value) / (max_value - min_value)
        normalized_values.append(normalized_value)

    return normalized_values
```

```python
df['Height'] = normalize(df['Height'])
print(df)
```
