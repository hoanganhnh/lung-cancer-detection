```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/Dataset pesticide and lung cancer.csv")
df.head()
```

```python
for i in range(df.shape[0]):
    if df.loc[i, 'ID'] % 1 == 0:
        df.loc[i, 'ID'] = True
    else:
        df.loc[i, 'ID'] = False
```

```python
col = df.pop('ID')
df.insert(len(df.columns), 'ID', col)
print(df)
```

```python
df.rename(columns={'ID': 'class'})
```

```python
df.to_csv('Pre Processing/1.1.reformatClass.csv', index=False)
```
